# GeoGram: IJP ÚJČ klasifikace singulár vs. plurál

Tento notebook testuje extrakci gramatického čísla z hesel na prirucka.ujc.cas.cz.

```python
from geogram.ujc import fetch_ujc_number

terms = ["Mohelnice", "Pardubice", "Lukavice", "Chornice"]

results = []
for term in terms:
    number = fetch_ujc_number(term)
    results.append({"term": term, "number": number})

import pandas as pd
pd.DataFrame(results)
```
