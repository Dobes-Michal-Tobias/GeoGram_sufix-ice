# GeoGram: filtrace obcí končících na -ice

Tento notebook ukazuje první krok: načtení seznamu obcí a výběr názvů končících na `-ice`.

```python
from pathlib import Path
import pandas as pd
from geogram.data import load_municipalities, filter_suffix_ice

data_path = Path("..ata/raw/municipalities.csv")
df = load_municipalities(data_path)
df_ice = filter_suffix_ice(df, name_column="name")

print(f"Total rows: {len(df):,}")
print(f"-ice rows: {len(df_ice):,}")
df_ice.head()
```
