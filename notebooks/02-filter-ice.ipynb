# GeoGram: filtrace obcí končících na -ice

Tento notebook ukazuje první krok: načtení seznamu obcí a výběr názvů končících na `-ice`.

```python
import sys
from pathlib import Path

project_root = Path('.').resolve().parent
sys.path.insert(0, str(project_root / 'src'))

from geogram.data import load_all_municipalities, filter_suffix_ice
import pandas as pd

data_dir = project_root / 'data' / 'raw'
df_all = load_all_municipalities(data_dir=data_dir)
df_ice = filter_suffix_ice(df_all, name_column='name')

print(f"Celkem obcí: {len(df_all):,}")
print(f"Obcí s koncovkou -ice: {len(df_ice):,}")
df_ice[['name', 'district_name', 'region_name']].head()
```
